```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('1ropani.csv')
# Remove the , and convert the price to an integer.
df['price'] = df['price'].apply(lambda x: int(x.replace(',','')))
```

```python

def clean_floor(x):
    x = x.lower()
    cleaned_floor = 0  # initialize cleaned floor value as zero
    # first getting the values after the decimal point
    if x == 'to and half floors': # cleaning a rare case wherre x.75 floor
        return 2.5
    if '.' in x:
        # y contains points after the decimal
        y = int(x.split(' ')[0].split('.')[1])
        if y == 75:
            cleaned_floor = 0
            return cleaned_floor
    
        cleaned_floor = cleaned_floor + (y / 10)
        z = int(x.split(' ')[0].split('.')[0])
        cleaned_floor = cleaned_floor + z
        return cleaned_floor
    if 'half' in x:
        cleaned_floor = cleaned_floor + 0.5
    if 'one' in x or '1' in x:
        cleaned_floor = cleaned_floor + 1
        
    if 'two'in x or '2' in x:
        cleaned_floor = cleaned_floor + 2
    if 'three'in x or '3' in x:
        cleaned_floor = cleaned_floor + 3
    if 'four'in x or '4' in x:
        cleaned_floor = cleaned_floor + 4
        
    if 'five'in x or '5' in x or 'fifth' in x:
        cleaned_floor = cleaned_floor + 5

    if 'six' in x or '6' in x:
        cleaned_floor = cleaned_floor + 6

    if 'seven'in x or '7' in x:
        cleaned_floor = cleaned_floor + 7

    if 'eight' in x or '8' in x:
        cleaned_floor = cleaned_floor + 8
    return cleaned_floor
  
    
df['floor'] = df['floor'].apply(clean_floor)
```

```python

# Cleaning bedroom
df['bedroom'] = df['bedroom'].apply(lambda x: int(x.split(' ')[0]))
```

```python
# cleaning bathroom
df['bathroom'] = df['bathroom'].apply(lambda x: int(x.split(' ')[0]))
```

```python
# Cleaning living room
df['livingroom'] = df['livingroom'].apply(lambda x: int(x.split(' ')[0]))
```

```python
# cleaning kitchen
df['kitchen'] = df['kitchen'].apply(lambda x: int(x.split(' ')[0]))
```
