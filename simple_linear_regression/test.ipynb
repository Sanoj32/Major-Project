```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
plt.rcParams['figure.figsize'] =(12.0,9.0)

data = pd.read_csv('data.csv')

X = data.iloc[:,0]
Y = data.iloc[:,1]

plt.scatter(X,Y)
```

```python
# Initializing slopes and intercepts as zero
c = 0
m = 0
L = 0.0001 # learning rate low for higher accuracy
epochs = 1000  # Number of iterations to perform gradient descent
n = float(len(X))


Y_pred = m*X + c

for i in range(epochs):
    Y_pred = m*X + c
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative with respect to slope m
    D_c = (-2/n) * sum(Y - Y_pred)        # Derivative with respect to y-intercept c
    m = m - L * D_m                       # Update m accordiong to partial derivative and learning rate L
    c = c - L * D_c                       # Update c accordiong to partial derivative and learning rate L
print(m,c)

```

```python
# Making predictions
Y_pred = m*X + c
plt.scatter(X,Y)
plt.scatter(X,Y_pred)
```
